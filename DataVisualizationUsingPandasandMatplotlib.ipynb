```python
 import pandas as pd
```

```python
from matplotlib import pyplot as mpl 
```

```python
dataSample = pd.read_csv('WBGlobalWelfare.csv')
dataSampleRural = dataSample[(dataSample['country_code']=="IND") & (dataSample['percentile']==1) & (dataSample['reporting_level']=="rural")]
dataSampleRural = dataSampleRural.filter(items=["year","avg_welfare"])
dataSampleUrban = dataSample[(dataSample['country_code']=="IND") & (dataSample['percentile']==1) & (dataSample['reporting_level']=="urban")]
dataSampleUrban = dataSampleUrban.filter(items=["year","avg_welfare"])
```

```python
mpl.plot(dataSampleRural.year, dataSampleRural.avg_welfare, label = "Rural")
mpl.plot(dataSampleUrban.year, dataSampleUrban.avg_welfare, label = "Urban")
mpl.title("India Avg Welfare")
mpl.legend()
mpl.show()
```
